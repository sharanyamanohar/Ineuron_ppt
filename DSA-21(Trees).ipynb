You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def in_order_traversal(node, values):
    if node is None:
        return
    
    in_order_traversal(node.left, values)
    values.append(node.value)
    in_order_traversal(node.right, values)

def construct_binary_search_tree(values):
    if not values:
        return None

    mid = len(values) // 2
    root = TreeNode(values[mid])

    root.left = construct_binary_search_tree(values[:mid])
    root.right = construct_binary_search_tree(values[mid+1:])

    return root

def convert_binary_tree_to_bst(root):
    values = []
    in_order_traversal(root, values)
    values.sort()

    new_root = construct_binary_search_tree(values)
    return new_root

# Construct the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert the binary tree to a binary search tree
new_root = convert_binary_tree_to_bst(root)

# Print the binary search tree
def print_in_order(node):
    if node is None:
        return
    
    print_in_order(node.left)
    print(node.value, end=" ")
    print_in_order(node.right)

print("Output:")
print_in_order(new_root)
```

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

**Input-1:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

**Output-1:**

The distance between the two keys = 4

**Input-2:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

**Output-2:**

The distance between the two keys = 2

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return TreeNode(value)
    
    if value < root.value:
        root.left = insert(root.left, value)
    elif value > root.value:
        root.right = insert(root.right, value)
    
    return root

def find_distance(root, node1, node2):
    if root is None:
        return 0
    
    # If both nodes are smaller than the root, search in the left subtree
    if node1 < root.value and node2 < root.value:
        return find_distance(root.left, node1, node2)
    
    # If both nodes are greater than the root, search in the right subtree
    if node1 > root.value and node2 > root.value:
        return find_distance(root.right, node1, node2)
    
    # Nodes are in different subtrees, the common ancestor is found
    return distance_to_node(root, node1) + distance_to_node(root, node2)

def distance_to_node(root, node):
    if root is None:
        return 0
    
    if node < root.value:
        return 1 + distance_to_node(root.left, node)
    elif node > root.value:
        return 1 + distance_to_node(root.right, node)
    
    return 0

# Create the BST
values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
root = None
for value in values:
    root = insert(root, value)

# Example 1
node1 = 6
node2 = 14
distance = find_distance(root, node1, node2)
print("Output 1:")
print("The distance between the two keys =", distance)

node1 = 3
node2 = 4
distance = find_distance(root, node1, node2)
print("\nOutput 2:")
print("The distance between the two keys =", distance)
```

ðŸ’¡ Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def convert_to_doubly_linked_list(root):
    if root is None:
        return None
    
    # Convert the left subtree to a doubly linked list
    left_list = convert_to_doubly_linked_list(root.left)
    
    # Convert the right subtree to a doubly linked list
    right_list = convert_to_doubly_linked_list(root.right)
    
    # Create a new doubly linked list node for the current root
    new_node = TreeNode(root.value)
    
    # Merge the left list, current node, and right list
    new_node.prev = left_list
    if left_list:
        left_list.next = new_node
    
    new_node.next = right_list
    if right_list:
        right_list.prev = new_node
    
    # Find the head of the doubly linked list
    head = new_node
    while head.prev:
        head = head.prev
    
    return head

def print_doubly_linked_list(head):
    current = head
    while current:
        print(current.value, end=" ")
        current = current.next
    print()

# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert the binary tree to a doubly linked list
head = convert_to_doubly_linked_list(root)

print("Output:")
print_doubly_linked_list(head)
```

ðŸ’¡ Question-4:

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

4 â†’ 5

5 â†’ 6

6 â†’ 7

7 â†’ -1

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None  # Pointer to the next node at the same level

def connect_nodes_at_same_level(root):
    if root is None:
        return
    
    # Use a dummy node to keep track of the start of the next level
    dummy = TreeNode(-1)
    prev = dummy
    
    # Start with the root node
    current = root
    
    while current:
        # Process all nodes at the current level
        while current:
            # Connect the left child, if it exists
            if current.left:
                prev.next = current.left
                prev = prev.next
            
            # Connect the right child, if it exists
            if current.right:
                prev.next = current.right
                prev = prev.next
            
            # Move to the next node at the current level
            current = current.next
        
        # Move to the next level
        current = dummy.next
        dummy.next = None
        prev = dummy

def print_linked_list(root):
    current = root
    while current:
        if current.next:
            print(f"{current.value} ->", end=" ")
        else:
            print(f"{current.value} -> -1")
        current = current.next

# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Connect nodes at the same level
connect_nodes_at_same_level(root)

# Print the linked list representation
print("Output:")
print_linked_list(root)
```
