Given preorder of a binary tree, calculate itsÂ **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)**Â [starting from depth 0]. The preorder is given as a string with two possible characters.

1. â€˜lâ€™ denotes the leaf
2. â€˜nâ€™ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can â€˜nâ€™ or â€˜lâ€™ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :

```python
def calculateDepth(preorder):
    depth = 0
    stack = []

    for char in preorder:
        if char == 'n':
            depth += 1
            stack.append(char)
        elif char == 'l':
            if not stack or stack[-1] != 'n':
                
                return -1

            depth -= 1
            stack.pop()
            if not stack or stack[-1] != 'n':
                
                return -1
            stack.pop()

    if stack:
        
        return -1

    return depth

preorder = 'nlnll'
print(calculateDepth(preorder))
```

Given aÂ Binary tree, the task is to print theÂ **left view**Â of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          / Â  \

        5 Â  Â  2

             / Â  \

            3 Â  Â  1

           / Â \

          6 Â  Â 7

***Output:**Â 4 5 3 6*


```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def printLeftView(root):
    if root is None:
        return

    queue = deque()
    queue.append(root)

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.popleft()

            if i == 0:
                print(node.data, end=" ")

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)

printLeftView(root)
```

ðŸ’¡ Question-3:

Given a Binary Tree, print the Right view of it.

TheÂ right view of a Binary TreeÂ is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      / Â  Â  \

   2 Â  Â   Â  Â 3

/ Â  \ Â  Â  Â  / Â \

4 Â  Â  5 Â  6 Â  Â 7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def right_view(root):
    if not root:
        return []

    result = []
    queue = deque([(root, 0)])

    while queue:
        node, level = queue.popleft()

        if level == len(result):
            result.append(node.val)

        if node.right:
            queue.append((node.right, level + 1))

        if node.left:
            queue.append((node.left, level + 1))

    return result
root1 = Node(1)
root1.left = Node(2)
root1.right = Node(3)
root1.left.left = Node(4)
root1.left.right = Node(5)
root1.right.left = Node(6)
root1.right.right = Node(7)
root1.right.right.right = Node(8)

result1 = right_view(root1)
print("Right view of the tree is:", result1) 

root2 = Node(1)
root2.left = Node(8)
root2.left.left = Node(7)

result2 = right_view(root2)
print("Right view of the tree is:", result2)
```

Given a Binary Tree, The task is to print theÂ **bottom view**Â from left to right. A nodeÂ **x**Â is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â  Â  \

5 Â  Â  Â  Â  3 Â  Â  Â  Â 25

        / Â  Â \

   10 Â  Â  Â  14

**Output:**Â 5, 10, 3, 14, 25.

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â / Â  \

 5 Â  Â  Â  Â  3 Â   4 Â  Â  25

         / Â  Â \

     10 Â  Â  Â  14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def bottom_view(root):
    if not root:
        return []

    hd_map = {}  # dictionary to store horizontal distance and nodes

    queue = deque([(root, 0)])

    while queue:
        node, hd = queue.popleft()

        # Update the horizontal distance map with the current node
        hd_map[hd] = node.val

        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    # Retrieve the bottom view nodes from the horizontal distance map
    bottom_view_nodes = [node_val for hd, node_val in sorted(hd_map.items())]

    return bottom_view_nodes
# Example 1
# Create the binary tree
root1 = Node(20)
root1.left = Node(8)
root1.right = Node(22)
root1.left.left = Node(5)
root1.left.right = Node(3)
root1.right.right = Node(25)
root1.left.right.left = Node(10)
root1.left.right.right = Node(14)

# Get the bottom view of the tree
result1 = bottom_view(root1)
print("Bottom view of the tree is:", result1)  # Output: Bottom view of the tree is: [5, 10, 3, 14, 25]

# Example 2
# Create the binary tree
root2 = Node(20)
root2.left = Node(8)
root2.right = Node(22)
root2.left.left = Node(5)
root2.left.right = Node(3)
root2.right.left = Node(4)
root2.right.right = Node(25)
root2.left.right.left = Node(10)
root2.left.right.right = Node(14)

# Get the bottom view of the tree
result2 = bottom_view(root2)
print("Bottom view of the tree is:", result2)  # Output: Bottom view of the tree is: [5, 10, 4, 14, 25]
```
